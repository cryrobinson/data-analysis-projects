### 5. Data Selection - Based on Conditional Filtering

```python
# Use the code block to select movies that have been released between 2010-2016, with a rating of less than 6.0, but generated a signifianct (top) revenue.
data_indexed['Year'] = pd.to_datetime(data_indexed['Year'], format='%Y').dt.year
data_indexed['Rating'] = data_indexed['Rating'].astype(float)

filtered_movies = data_indexed[
    (data_indexed['Year'].between(2010, 2016)) &
    (data_indexed['Rating'] < 6.0)
]

top_revenue_movies = filtered_movies.sort_values(by='Revenue (Millions)', ascending=False).head(10)

print(top_revenue_movies)
```
